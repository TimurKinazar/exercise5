```
import numpy as np
import math
from math import cos,sin,sqrt
from matplotlib.patches import Circle
import matplotlib.pyplot as plt
theta = np.arange(0., 2., 1./180.)*np.pi
plt.polar(theta, [1.0]*len(theta))
n=100
R=np.random.uniform(low = 0, high = 1, size = n)
for i in range(100):
  R[i] = math.sqrt(R[i])
#квадратный корень извлекается для сохранения равномерности
phi=2*np.pi*np.random.uniform(low = 0, high = 1, size = n)
plt.polar(phi, (R), "o")
```
