```
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.sin(t), np.cos(t))
plt.axis('equal')
X = np.zeros(100)
Y = np.zeros(100)
k=0
#строим набор чисел, равномерно распределенных, и записываем только те,первые 100
#что попадают в круг. 
while k<100:
  x= (1-2*np.random.uniform(low =0, high = 1, size = 1))
  y= (1-2*np.random.uniform(low =0, high = 1, size = 1))
  ro2 = x**2+y**2
  if ro2 <= 1:
    X[k]=x
    Y[k]=y
    k+=1
X=X[1:k]
Y=Y[1:k]
plt.scatter(X,Y)
```
